# Transposing the data

```python
dataset=dataset.transpose()
```

```python
dataset.head(26)
```

# Missing data

```python
X=dataset.iloc[:,:].values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='constant',fill_value=1)
imputer.fit(X[1:, :])
X[1:, :] = imputer.transform(X[1:, :])
```

# Calculating CAGR(For finding Growth Of each State)

```python
CAGR=[0] #Compound Annual Growth Rate
for i in range(1,35):
    x=X[i][0]
    y=X[i][(len(X[i])-1)]
    C=((float(y)/float(x))**(1/len(X[i])))-1
    CAGR.append(C)
dataset['CAGR']=CAGR
```

```python
dataset.head(25)
```

```python
rows_name=[]
for name in dataset.index:
    rows_name.append(name)
rows_name.pop(0)
CAGR.pop(0)
plt.bar(rows_name,CAGR)
plt.xticks(rotation=90)
plt.show()
```

# Applying Support Vector Machine

```python
X=dataset.iloc[1:,:24].values
y=dataset.iloc[1:,[25]].values
```

```python
#We have set a Limit in this case i.e 0.4 that State having CAGR more than 0.3 are develeped and less than 0.3 are Less developed
```

```python
z=[]
for data in y:
    if data >0.04:
        z.append(1)
    else:
        z.append(0)
print(z)
print(z.count(0))
print(z.count(1))
```

# Spiliting the data into test set and train set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, z, test_size = 0.25, random_state = 0)
```

# Training the SVC model

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```


# Predicting the Result

```python
y_pred = classifier.predict(X_test)
print(y_pred)
print(y_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
